## Exercise 2
Using scatter plot:  
a) Show the relation of square footage and price.  
b) Check how price is influenced by apartment's grade. Find other variables correlated with price.  
c) Check how good apartments are distributed over the city. First, plot zipcode versus price. Then, use apartments coordinates (longitude and lattitude) in order to show where apartments of good grade are located. 

```python
#a)
#Your code:
plt.scatter(x = data['sqft_living'], y = data['price'])
plt.ticklabel_format(axis='both', style='plain')
plt.xlabel('sqft living'), plt.ylabel('price') 

#Hint plt.scatter
```

Here we can see the ratio of price to square meters of the house. The cheapest houses have sq footage between 1000 to 3000 and their price is between around $500.000 - $1.500.000.

```python
#b)
#Your code:
plt.scatter(x=data['grade'], y=data['price'])
plt.ticklabel_format(axis='both', style='plain')
plt.xlabel('grade'), plt.ylabel('price') 
```

As expected, the higher the class of the apartment, the bigger the price. As the grades of the houses hit the extremes (min and max) we can see that their number drastically declines.

```python
#c)
#Your code:
plt.scatter(x=data['zipcode'], y=data['price'])
plt.ticklabel_format(axis='both', style='plain')
plt.xlabel('zipcode'), plt.ylabel('price') 
```

Here we can see the ratio of price to zip code of the house.

```python
#Your code:
plt.scatter(x = data['long'], y = data['lat'], c = data['grade'], s = 25)
plt.ylabel('lattitude'), plt.xlabel('longitude')
```

In this graph we can see how good apartments are distributed over the city. Using apartments coordinates (longitude and lattitude) to show where apartments of good grade are located. 

## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
#Your code:
data.boxplot(column='price', by='grade')
plt.title(''), plt.suptitle(''), plt.ylabel('price')
plt.ticklabel_format(axis='y', style='plain')
plt.grid(b=None)
#Hint: plt.boxplot
```

This boxplot is useful to check the outliers in prices. As we can see, as the house grade increases, its price also increases. Also the variability in price increases as house grade increases.


```python
#b)
from statsmodels.graphics.mosaicplot import mosaic

#Your code:
mosaic(data, ['bedrooms', 'floors'])
```

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
sort_rooms = data[(data['bathrooms'] == 2) & (data['bedrooms'] == 3) & (data['floors'] == 2)]
sort_rooms.drop(['date'], axis = 1).describe(include = 'all')
#Hint: pandas.DataFrame.loc
```

We can analyze this output, like in a task 1.c) using fulfilling above constraints.

```python
#b)
#Your code:
sort_rooms = data[(data['bathrooms'] == 2) & (data['bedrooms'] == 3) & (data['floors'] == 2)]
plt.scatter(x=data['grade'], y=data['price'], s = 5, color = 'black')
plt.ticklabel_format(axis='both', style='plain')
plt.xlabel('grade'), plt.ylabel('price') 
plt.scatter(x=sort_rooms['grade'], y=sort_rooms['price'], s = 15, alpha=0.75, color = 'y')
plt.rcParams["figure.figsize"] = (15,10)
```

Here we can see, that the most houses are in good grades and are not close to hitting the given top prices

```python
#c)
#Your code:
sort_rooms = data[(data['bathrooms'] == 2) & (data['bedrooms'] == 3) & (data['floors'] == 2)]
sort_rooms.loc[(sort_rooms['price'] / sort_rooms['sqft_living']).sort_values().index].head()

#Hint: .loc, .sort_values(), .index
```

In this case we found the best price to footage ratio for 5 apartments.

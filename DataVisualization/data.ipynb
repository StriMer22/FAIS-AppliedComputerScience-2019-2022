```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sbu
```

```python
df = pd.read_csv("280221_metrics_real.tsv", sep='\t', header=0)

plt.figure()
fig, axes = plt.subplots(nrows=2, ncols=2, constrained_layout=True)
for temp, x, y in [['Total', 0, 0], ['Minimum', 0, 1], ['Maximum', 1, 0], ['Average', 1, 1]]:
    temp_df = df[[
        'TOI', 
        temp+'_duration_of_fixations.Prawa',
        temp+'_duration_of_fixations.Przd',
        temp+'_duration_of_fixations.Lewa',
        temp+'_duration_of_fixations.Tory'
    ]].loc[df['TOI'] != 'Entire Recording']
    temp_df.columns = ['TOI', 'prawa', 'przd', 'lewa', 'tory']
    temp_df.plot.hist(stacked=True, bins=50, ax=axes[x, y], title=temp)
    axes[x, y].legend(temp_df.columns[1:], fontsize='small')
plt.setp(axes[:, :], xlabel='Time in ms')
```

```python
df = pd.read_csv("280221_metrics_vr.tsv", sep='\t', header=0)
cmap = plt.cm.get_cmap('nipy_spectral')

from pandas.plotting import radviz
w = 1
m = '.'
plt.figure()
for temp, clr in [['Total', cmap(0)], ['Minimum', cmap(0.25)], ['Maximum', cmap(0.5)], ['Average', cmap(0.75)]]:
    temp_df = df[[
        'TOI', 
        temp+'_duration_of_fixations.Prawa',
        temp+'_duration_of_fixations.Przd',
        temp+'_duration_of_fixations.Lewa',
        temp+'_duration_of_fixations.Tory'
    ]].loc[df['TOI'] != 'Entire Recording']
    temp_df.columns = ['TOI', 'prawa', 'przd', 'lewa', 'tory']
    temp_df['TOI'] = temp+' duration'
    radviz(temp_df, 'TOI', color=clr, marker=m, linewidths=w)
```
